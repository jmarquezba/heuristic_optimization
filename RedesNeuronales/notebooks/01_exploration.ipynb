Visualiza las curvas de nivel de las seis funciones objetivo y verifica sus óptimos.
Forma parte del Trabajo 1 (IRNA 2026-01).

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib.pyplot as plt
from src.optimization.functions import ALL_FUNCTIONS, get_function
```

```python
# Verificación de óptimos conocidos
for name in ALL_FUNCTIONS:
    spec = get_function(name, 2)
    print(f'{name:18s} f(x*) = {float(spec.func(spec.optimum)):.6f}  (esperado {spec.f_opt})')
```

```python
# Curvas de nivel
fig, axes = plt.subplots(2, 3, figsize=(15, 9))
for ax, name in zip(axes.ravel(), ALL_FUNCTIONS):
    spec = get_function(name, 2)
    xs = np.linspace(spec.lower[0], spec.upper[0], 200)
    ys = np.linspace(spec.lower[1], spec.upper[1], 200)
    X, Y = np.meshgrid(xs, ys)
    Z = spec.func(np.column_stack([X.ravel(), Y.ravel()])).reshape(X.shape)
    ax.contourf(X, Y, Z, levels=30, cmap='viridis')
    ax.plot(spec.optimum[0], spec.optimum[1], 'r*', ms=12)
    ax.set_title(name)
plt.tight_layout(); plt.show()
```
