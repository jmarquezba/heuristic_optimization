Corre una versión rápida de los experimentos (o carga `numerical_results.csv`) y

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import pandas as pd
from src.utils.config import load_config, RESULTS_DIR
from src.optimization import experiment_runner
```

```python
# Demo rápida (subconjunto). Ajusta o usa run_all.py para el experimento completo.
raw = load_config('numerical_experiments.yaml')
cfg = {k: v for k, v in raw.items() if k not in ('fast', 'full')}
cfg.update(raw['fast'])
cfg['functions'] = ['rastrigin', 'rosenbrock']
cfg['dims'] = [2]
cfg['gradient_descent']['n_runs'] = [30]
df, summary = experiment_runner.run_all(cfg)
summary
```

```python
# Comparación de valor final por algoritmo
import matplotlib.pyplot as plt
for fn in cfg['functions']:
    sub = df[(df.function == fn) & (df.dim == 2)]
    algs = sorted(sub.algorithm.unique())
    plt.figure(figsize=(7,4))
    plt.boxplot([sub[sub.algorithm==a]['f_best'] for a in algs], labels=algs)
    plt.title(f'Valor final · {fn} 2D'); plt.xticks(rotation=20); plt.grid(alpha=0.3)
    plt.tight_layout(); plt.show()
```
