Construye los datos, el modelo de costos y corre ACO/GA (versión rápida).
Para resultados completos y animaciones: `python run_all.py --only tsp --mode full`.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import matplotlib.pyplot as plt
from src.tsp.data_loader import build_dataset
from src.tsp.cost_model import Vehicle, CostParams, build_cost_matrices, route_breakdown
from src.tsp.ant_colony import ant_colony_tsp
from src.tsp.genetic_algorithm import genetic_algorithm_tsp
from src.visualization.maps import plot_route_comparison
```

```python
df = build_dataset(write=True)
mats = build_cost_matrices(df, Vehicle(), CostParams(), valor_hora=30.0)
cost = mats['total']
print('Ciudades:', len(df), '| matriz:', cost.shape)
df.head()
```

```python
aco = ant_colony_tsp(cost, seed=1983, n_ants=20, n_iter=80)
ga = genetic_algorithm_tsp(cost, seed=1983, pop_size=120, n_gen=200)
print('ACO costo:', round(aco.best_cost), '| GA costo:', round(ga.best_cost))
plot_route_comparison(df, aco.best_route, ga.best_route, out_path='_tmp_cmp.png')
plt.figure(figsize=(8,4))
plt.plot(aco.history, label='ACO'); plt.plot(ga.history, label='GA')
plt.xlabel('iteración'); plt.ylabel('mejor costo (€)'); plt.legend(); plt.grid(alpha=0.3)
plt.title('Convergencia TSP (valor hora = 30 €/h)'); plt.show()
```
